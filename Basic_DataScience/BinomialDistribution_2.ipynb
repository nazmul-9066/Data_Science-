```python
import scipy.stats as st
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline


plt.figure(figsize = (10, 8))

sucessesRate = np.linspace(10, 90, 80) # (starting, ending, datapoint)

plt.plot(sucessesRate, st.binom.pmf(sucessesRate, 100, 0.4), 'r-', label = "Binom (100, 0.4)" ) # pmf = probability mass function

higher_sucessesRate = sucessesRate[sucessesRate > 85]

plt.fill_between(higher_sucessesRate, 0, st.binom.pmf(higher_sucessesRate, 100, 0.4), alpha = 0.7, label = "P-value")

_ = plt.legend()

Pvalue = st.binom.cdf(85, 100, 0.4)

print(Pvalue)

plt.show()


```
