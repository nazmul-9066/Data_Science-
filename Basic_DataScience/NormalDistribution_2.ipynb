```python
import numpy as np 
import matplotlib.pyplot as plt
import scipy.stats as st
%matplotlib inline

# why we use in inline?

meanvalues = 5
variance = 14
std = np.sqrt(variance)

linesplot = np.linspace(meanvalues - 5*variance, meanvalues + 5*variance, 100)

plt.figure(figsize = (8,4))
plt.plot(linesplot, st.norm.pdf(linesplot, meanvalues, std)) # probability density function
```
